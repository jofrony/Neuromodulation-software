```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_ACh_pause_control(obj_loaded_traces,type_cell='dSPN'):
    data_dict = obj_loaded_traces.get_traces(neuronType=type_cell)
    
   
    
    control = np.array([*data_dict[5].values()])

    da = np.array([*data_dict[1].values()])
    
    control_m = np.mean(control,axis=0)
    control_std = np.std(control,axis=0)
    
    da_m = np.mean(da,axis=0)
    da_std = np.std(da,axis=0)
    
    
    time = np.arange(0,3000.5,0.5)
    for i in np.arange(1,10,1):
            fig = plt.figure()
            ax = fig.add_subplot(111)
            ax.plot(time,control[i]*1e3,c='black')
            ax.plot(time,da[i]*1e3,c='red')
            ax.set_ylabel('Membrane potential (mV)')
            ax.set_xlabel('Time(ms)')
            
            ax.spines['right'].set_visible(False)
            ax.spines['top'].set_visible(False)
            ax.yaxis.set_ticks_position('left')
            ax.xaxis.set_ticks_position('bottom')
            ax.plot(np.arange(2500,3000),np.ones_like(np.arange(2500,3000))*-75,c='black')
            ax.text(2500,-74,'500 ms')
            ax.plot(2500*np.ones_like(np.arange(-80,-75)),np.arange(-80,-75),c='black')
            ax.text(2550,-78,'5 mV')
            ax.set_ylim([-100,-70])
            
            fig.savefig('Level1/dopamine-control-traces'+str(i)+".svg")
            
    plt.figure()
    
    plt.plot(time,control_m,c='black')
    plt.plot(time,da_m,c='red')
    plt.fill_between(time,control_m-control_std,control_m+control_std,alpha=.1,color='black')
    plt.fill_between(time,da_m-da_std,da_m+da_std,alpha=.1,color='red')      
    
    plt.show()
                
    
    
```

```python
plot_ACh_pause_control(kl,type_cell='dSPN')
```

```python
def plot_dopamine_control(obj_loaded_traces,type_cell='dSPN'):
    data_dict = obj_loaded_traces.get_traces(neuronType=type_cell)
    
   
    
    control = np.array([*data_dict[5].values()])

    da = np.array([*data_dict[4].values()])
    
    control_m = np.mean(control,axis=0)
    control_std = np.std(control,axis=0)
    
    da_m = np.mean(da,axis=0)
    da_std = np.std(da,axis=0)
    
    
    time = np.arange(0,3000.5,0.5)
    for i in np.arange(1,10,1):
            fig = plt.figure()
            ax = fig.add_subplot(111)
            ax.plot(time,control[i]*1e3,c='black')
            ax.plot(time,da[i]*1e3,c='red')
            ax.set_ylabel('Membrane potential (mV)')
            ax.set_xlabel('Time(ms)')
            
            ax.spines['right'].set_visible(False)
            ax.spines['top'].set_visible(False)
            ax.yaxis.set_ticks_position('left')
            ax.xaxis.set_ticks_position('bottom')
            ax.plot(np.arange(2500,3000),np.ones_like(np.arange(2500,3000))*-75,c='black')
            ax.text(2500,-74,'500 ms')
            ax.plot(2500*np.ones_like(np.arange(-80,-75)),np.arange(-80,-75),c='black')
            ax.text(2550,-78,'5 mV')
            ax.set_ylim([-100,-70])
            
            fig.savefig('Level1/dopamine-control-traces'+str(i)+".svg")
            
    plt.figure(0)
    
    plt.plot(time,control_m,c='black')
    plt.plot(time,da_m,c='red')
    plt.fill_between(time,control_m-control_std,control_m+control_std,alpha=.1,color='black')
    plt.fill_between(time,da_m-da_std,da_m+da_std,alpha=.1,color='red')      
    
    plt.show()
                
    
    
```

```python
plot_dopamine_control(kl,type_cell='dSPN')
```

```python
def depolarisation_response(obj_loaded_traces,type_cell,start=1,end=1.5):
    data_dict = obj_loaded_traces.get_traces(neuronType=type_cell)
    depol = dict()
    for k, data in data_dict.items():
        for t, voltages in data.items():
            depolarisation = list()
            if max(voltages) < 0:
                depolarisation.append(abs(voltages[int(start/5e-4):int(end/5e-4)]) - abs(data_dict[5][t][int(start/5e-4) : int(end/5e-4)]))
                
        depol.update({k : depolarisation})
                
    return depol
```

```python
def plot_box_whisker(depol,chosen=[4]):

    for k, session_data in depol.items():
        
        if k in chosen:
            fig = plt.figure()
            ax = fig.add_subplot(111)
            corrected = np.array(session_data)*1e3
            ax.boxplot(corrected[0])
            ax.set_ylabel('Membrane potential change (mV)')
            ax.spines['right'].set_visible(False)
            ax.spines['top'].set_visible(False)
            ax.yaxis.set_ticks_position('left')
            ax.xaxis.set_ticks_position('bottom')
    plt.tight_layout()
    plt.xlim([0.5,1.5])
    plt.savefig('Level1/box_whisker.svg')
       
    plt.show()
```

```python
def mean_depolarisation(depol):
    average = dict()
    for k, session_data in depol.items():
        corrected = np.array(session_data)*1e3
        mean = corrected.mean()
        std = corrected.std()
        average.update({k : {'mean' : mean, 'std' : std}})
       
    return average
```

```python
def plot_membrane_potential(mean_depol,labels_v=['v1','v2'],exp_title=None,chosen=[4]):
    fig, ax = plt.subplots()
    pos = 0
    for k, average in mean_depol.items():
        if k in chosen:
            ax.bar([pos], [average['mean']],yerr=[average['std']], align='center', alpha=0.5, ecolor='black', capsize=5,width=0.4)
            ax.scatter([pos], [average['mean']])
            pos=pos+1
    ax.set_ylabel('Membrane potential change (mV)')
    ax.set_title('Membrane potential change (mV)')
    ax.set_xticks(np.arange(len(chosen)))
    ax.set_xticklabels(labels_v)
    ax.set_xlim([-0.5,0.5])
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')


    plt.tight_layout()
    plt.title(exp_title)
    plt.savefig('Level1/' + exp_title+'_mean_freuquency.svg')
    plt.show()
    
```

```python
depol_dSPN = depolarisation_response(kl,type_cell='dSPN')
mdepoldSPN = mean_depolarisation(depol_dSPN)
```

```python
plot_box_whisker(depol_dSPN)
```

```python
plot_box_whisker(depol_dSPN,chosen=[0,1,2,3,4])
```

```python
plot_membrane_potential(mdepoldSPN ,labels_v=['DA'],exp_title='dSPN depol')
```
