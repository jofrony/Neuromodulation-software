# Dopaminergic modulation of iSPN

```python
from ispn_analysis import iSPNanalysis
from neuromodulation.experimental_data import experimental
```

```python
dl = iSPNanalysis('pdc/iSPN/ispn/str-ispn-e150908_c4_D2-m51-5-DE-v20210215/dopamine/ID_0/')
dl.load()
dl.dopamine_analysis()


iSPNmodulation = experimental()
iSPNmodulation.define_exp(control = { "mean" : dl.get_dopamine_control()[0], "std" : 0})
iSPNmodulation.define_exp(modulated = { "mean" : dl.get_dopamine_control()[0] - dl.get_criteria()['mean'], \
                                       "std" : dl.get_criteria()['std'] * 1.5})
iSPNmodulation.define_exp(control_sim = dl.get_dopamine_control())
iSPNmodulation.define_exp(modulated_sim = [dl.num_AP_passing_analysis_traces_dopamine])
dl.plot_comparison(iSPNmodulation.experimental_data['control'],iSPNmodulation.experimental_data['control_sim'],\
                   iSPNmodulation.experimental_data['modulated'], iSPNmodulation.experimental_data['modulated_sim'],\
                   x_ticks=['Model 1'],ylabel='Number of APs',title='Dopamine modulation',num_models=1,save=True,\
                   filename='model_1.pdf' )
```

```python
dl.save_selected_modulation_dopamine()
modulation, voltage = dl.get_final_modulation()

dl.plot_chosen_modulation()
```
