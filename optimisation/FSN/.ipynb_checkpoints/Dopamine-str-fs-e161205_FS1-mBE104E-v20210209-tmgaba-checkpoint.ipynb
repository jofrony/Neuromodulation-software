### Set the modulation : Dopamine ( D2 receptor) gabaergic input

```python
from neuromodulation.modulation_set import defineModulation
```

#### parameter ID : 0

```python
cellTypefolder = '../../single_cell_models/HBP-2021Q1/FSN/str-fs-e161205_FS1-mBE104E-v20210209/'
out_put_dir = 'fsn/str-fs-e161205_FS1-mBE104E-v20210209/tmgaba'
parameterID = 0
ModSet = defineModulation(parameterID = parameterID)
ModSet.cell_name('FSN')
ModSet.define_neuromodulation(modulationA = "dopamine")
ModSet.new_modulation_dir(out_put_dir)
ModSet.tstop = 1000
ModSet.set_time_step(dt=0.025)
ModSet.population = 4
ModSet.cellDir = cellTypefolder

ModSet.define_modulation_parameter('kir',"dopamine",'somatic', [0.7,0.9])
ModSet.define_modulation_parameter('kir',"dopamine",'basal', [0.7,0.9])
ModSet.define_modulation_parameter('kas',"dopamine",'somatic', [0.7,1.0])
ModSet.define_modulation_parameter('kaf',"dopamine",'somatic', [0.95,1.05])
ModSet.define_modulation_parameter('naf',"dopamine",'somatic', [0.95,1.05])


## NEURON tau in ms
ModSet.define_modulation_receptor('tmGabaA','dopamine', {'tau1' : 1.2, 'tau2' : 8},\
                                  {"maxMod" : [0.5,0.7]})

ModSet.define_protocol(typeEx = 'synaptic_input', parameters = {"conductance" : 1e-3, "delay": 0, \
                                                                "threshold": 0, \
                                                                "spiketimes" : [300,600]})
ModSet.define_modulation_function(modulation_function = {"function" : "bath_application","gmax" : 1})
ModSet.define_protocol(typeEx = 'current_clamp', parameters = {"start" : 0,"duration" : 1000 , "amp" : 0.1})

ModSet.define_selection_criteria(function = "synaptic_amplitude_decrease_percentage", criteria ={ "selection" : {"mean" : 60, "std" : 20, "threshold":1.5}, "parameters" : {'start_base' : 50, 'stop_base': 100, 'start_measure' : 300, 'stop_measure' : 400}})
ModSet.save_modulation_setup()
ModSet.save_modulation()
ModSet.neuromodulationDir
```
