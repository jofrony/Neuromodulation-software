```python
import numpy as np
```

```python
data = np.loadtxt("fs/str-fs-e161024_FS16-mDR-rat-Mar-13-08-1-536-R-v20190225/acetylcholine/voltages.txt")
```

```python
import neuromodulation.modulation_functions as mf 
modulation = {"function" : "bath_application","gmax" : 1, "dt": 0.025,"tstop":1000,"ht" : np.arange(0,700,0.025)}
time_series = mf.bath_application(modulation)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,6))
plt.subplot(131)
for k in data[5:]:
    plt.plot(data[0][9000:],k[9000:])
    
plt.plot(data[0][9000:],data[-1][9000:],label="Control")
plt.title("D1 receptor modulation")
plt.ylabel("membrane potential (mV)")
plt.xlabel("Time (ms)")
plt.legend()

plt.subplot(132)
plt.plot(modulation["ht"][9000:],time_series[9000:])
plt.title("D1 receptor modulation")
plt.xlabel("Time (ms)")
```
