# Muscarinic modulation of dSPN

```python
from dspn_analysis import dSPNanalysis
from neuromodulation.experimental_data import experimental
```

### ID 0

```python
from dspn_analysis import dSPNanalysis

dl = dSPNanalysis('pdc/dSPN/dspn/str-dspn-e150917_c10_D1-mWT-P270-20-v20210212/acetylcholine/ID_0/')
dl.load()
dl.muscarinic_analysis()


dSPNmodulation = experimental()
dSPNmodulation.define_exp(control = { "mean" : dl.get_muscarinic_control()[0], "std" : 0})
dSPNmodulation.define_exp(modulated = { "mean" : dl.get_muscarinic_control()[0] + dl.get_criteria()['mean'], \
                                       "std" : dl.get_criteria()['std'] * 1.5})
dSPNmodulation.define_exp(control_sim = dl.get_muscarinic_control())
dSPNmodulation.define_exp(modulated_sim = [dl.num_AP_passing_analysis_traces_muscarinic])
dl.plot_comparison(dSPNmodulation.experimental_data['control'],dSPNmodulation.experimental_data['control_sim'],\
                   dSPNmodulation.experimental_data['modulated'], dSPNmodulation.experimental_data['modulated_sim'],\
                   x_ticks=['Model 1'],ylabel='Number of APs',title='Dopamine modulation',num_models=1,save=True,\
                   filename='model_1.pdf' )
```

```python
dl.save_selected_modulation()
modulation, voltage = dl.get_final_modulation()

dl.plot_chosen_modulation()
```
