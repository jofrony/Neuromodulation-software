```python
from neuromodulation import utils
import pathlib
import os
import shutil
import json
import glob
```

```python
snudda = pathlib.Path(os.path.abspath('../../../Snudda/snudda/data/synapses/striatum/'))
```

```python
fs = pathlib.Path(os.path.abspath('../optimisation/FS/fs/str-fs-e160628_FS2-mMTC180800A-IDB-v20190226/acetylcholine/receptor-modulations.json'))
```

```python
fs_load = json.load(open(fs,'rb'))
maxModB = fs_load[0][0][1]['value']
failRateB = fs_load[0][0][0]['value']
```

```python
for f in snudda.iterdir():
    if 'tmgaba' in os.path.abspath(f) and 'modulation' not in os.path.abspath(f):
        json_load = json.load(open(f,'rb'))
        
        for num, param in json_load.items():
            json_load[num]["synapse"].update({'maxModB' : maxModB, 'failRateB' : failRateB})

        name = os.path.abspath(f.parent / '.'.join([f.name.split('.')[0]+ '-modulation', f.name.split('.')[1]]))
        json.dump(json_load,open(name,'w'))
```
